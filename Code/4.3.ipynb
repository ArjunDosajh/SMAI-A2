```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
import numpy as np
from sklearn.mixture import GaussianMixture
```

```python
dataset = pickle.load(open('SMAI-Dataset-problem-4.3/colors.pkl', 'rb'))
dataset = np.array(dataset)
dataset[0]
```

## Visualizing the dataset

```python
# plot a np array of shape (100, 2) as a scatter plot
plt.scatter(dataset[:,0], dataset[:,1])
plt.show()
```

```python
gmm = GaussianMixture(n_components=7).fit(dataset)
labels = gmm.predict(dataset) # np array of labels of each point in the dataset
means = gmm.means_ # np array of mean of each cluster
covariances = gmm.covariances_ # np array of covariance of each cluster
```

```python
def generate_data(n, means, covariances):
    sample_dataset = []
    for i in range(n):
        samples = np.random.multivariate_normal(means[i], covariances[i], size=int(100/n))
        sample_dataset.append(samples)
    return np.vstack(sample_dataset)
```

## Scatter plot for the generated dataset

```python
generated_dataset = generate_data(7, means, covariances)
plt.scatter(generated_dataset[:,0], generated_dataset[:,1])
plt.show()

print(f"Generated Dataset:\n{generated_dataset[:10]}\nOriginal Dataset:\n{dataset[:10]}")
```
